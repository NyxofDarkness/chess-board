```python
class ChessBoard:
    """[Make a chessboard render with two queens present]
    """
    def __init__(self):
        self.grid = np.zeros((8, 8, 3))
        self.red = (1, 0, 0)
        self.blue = (0, 0, 1)
        self.white = (1, 1, 1)
        self.red_queen = tuple()
        self.blue_queen = tuple()
        self.board()

    def board(self):
        """[this finds the range and length of the whole grid, any squares that evenly divide by two are white, renders position of two queens]
        """
        for x in range(len(self.grid)):
            for y in range(len(self.grid[x])):
                if (x + y) % 2 == 0:
                    self.grid[x, y] = self.white
        if self.red_queen:
            self.grid[self.red_queen] = self.red
        if self.blue_queen:
            self.grid[self.blue_queen] = self.blue

    def add_red(self, row, col):
        """[creates position of red queen]
        """
        self.red_queen = (row, col)
        self.board()
        

    def add_blue(self, row, col):
        """[creates position of green queen]
        """
        self.blue_queen = (row, col)
        self.board()

    def render(self):
        """[renders board]
        """
        plt.imshow(self.grid)

    def is_under_attack(self):
        """[return boolean indicating if a queen is in striking position]
        """
        if self.red_queen[0] == self.blue_queen[0] or self.red_queen[1] == self.blue_queen[1]:
            return True
        elif abs(self.red_queen[0] - self.blue_queen[0]) == abs(self.red_queen[1] - self.blue_queen[1]):
            return True
        else:
            return False
```

```python
chess = ChessBoard()
chess.add_red(2, 4)
chess.add_blue(3, 1)
chess.render()
expected = False
actual = chess.is_under_attack()
print(actual)
assert actual == expected
```
